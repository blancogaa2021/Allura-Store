#1. Análisis de facturación



```python
import pandas as pd

# Cantidad de ventas realizadas
venta = len(tienda1)
venta2 = len(tienda2)
venta3 = len(tienda3)
venta4 = len(tienda4)
# Procederemos a calcular el ingreso bruto de cada tienda
Ingresobruto = tienda1["Precio"].sum()
Ingresobruto2 = tienda2["Precio"].sum()
Ingresobruto3 = tienda3["Precio"].sum()
Ingresobruto4 = tienda4["Precio"].sum()
# Procederemos a calcular el gastos incurridos por los envíos
gasto = tienda1["Costo de envío"].sum()
gasto2 = tienda2["Costo de envío"].sum()
gasto3 = tienda3["Costo de envío"].sum()
gasto4 = tienda4["Costo de envío"].sum()
# Procederemos a calcular los ingresos netos que son iguales a los ingresos menos los gastos
ingresoneto = Ingresobruto - gasto
ingresoneto2 = Ingresobruto2 - gasto2
ingresoneto3 = Ingresobruto3 - gasto3
ingresoneto4 = Ingresobruto4 - gasto4
# Procederemos a calcular los ingresos netos promedio por venta realizada
ingresopromedio = ingresoneto / len(tienda1)
ingresopromedio2 = ingresoneto2 / len(tienda2)
ingresopromedio3 = ingresoneto3 / len(tienda3)
ingresopromedio4 = ingresoneto4 / len(tienda4)
# Procederemos a calcular la media de califaciones de los clientes
calificacion = tienda1["Calificación"].mean()
calificacion2 = tienda2["Calificación"].mean()
calificacion3 = tienda3["Calificación"].mean()
calificacion4 = tienda4["Calificación"].mean()
# Procederemos a calcular la media de los envios por tienda
gastopromedio1 = tienda1["Costo de envío"].mean()
gastopromedio2 = tienda2["Costo de envío"].mean()
gastopromedio3 = tienda3["Costo de envío"].mean()
gastopromedio4 = tienda4["Costo de envío"].mean()


dicti = {"Tienda": ["1", "2", "3", "4"],
         "Cantidad de Ventas": [venta, venta2, venta3, venta4],
         "Ingreso bruto": [f"{Ingresobruto:,.2f}", f"{Ingresobruto2:,.2f}", f"{Ingresobruto3:,.2f}", f"{Ingresobruto4:,.2f}"],
         "Gasto": [f"{gasto:,.2f}", f"{gasto2:,.2f}", f"{gasto3:,.2f}", f"{gasto4:,.2f}"],
         "Ingreso neto": [f"{ingresoneto:,.2f}", f"{ingresoneto2:,.2f}", f"{ingresoneto3:,.2f}", f"{ingresoneto4:,.2f}"],
         "Ingreso medio": [f"{ingresopromedio:,.2f}", f"{ingresopromedio2:,.2f}", f"{ingresopromedio3:,.2f}", f"{ingresopromedio4:,.2f}" ],
         "Calificación": [f"{calificacion:,.2f}", f"{calificacion2:,.2f}", f"{calificacion3:,.2f}", f"{calificacion4:,.2f}"],
         "Gasto medio": [f"{gastopromedio1:,.2f}", f"{gastopromedio2:,.2f}", f"{gastopromedio3:,.2f}", f"{gastopromedio4:,.2f}"]
         }

df = pd.DataFrame(dicti)
df
```

```python
# Vamos a totalizar las cantidades y las ventas por categoria para poder identificar, que categoria aporta más en las ventas
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame consolidado
data_list = []
for i in range(1, 5):
    tienda_name = f"tienda{i}"
    tienda_df = globals()[tienda_name].copy()
    tienda_df['Tienda'] = tienda_name
    data_list.append(tienda_df)

df_consolidado = pd.concat(data_list)

# Totalizar cantidades y ventas por categoría
totales_por_categoria = df_consolidado.groupby('Categoría del Producto').agg(
    Cantidad=('Categoría del Producto', 'size'),
    Ingresos=('Precio', 'sum')
).reset_index()

# Mostrar la tabla de totales por categoría
print("\n--- Totales por Categoría ---")
totales_por_categoria['Ingresos_Formateado'] = totales_por_categoria['Ingresos'].map('${:,.2f}'.format)
print(totales_por_categoria[['Categoría del Producto', 'Cantidad', 'Ingresos_Formateado']])

# --- Gráficos de torta por categoría con leyenda a la derecha ---

# Gráfico de torta de cantidades por categoría
plt.figure(figsize=(10, 6))
plt.pie(totales_por_categoria['Cantidad'], autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Distribución de Cantidad de Ventas por Categoría')
plt.axis('equal')
plt.legend(totales_por_categoria['Categoría del Producto'], loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()

# Gráfico de torta de ingresos por categoría
plt.figure(figsize=(10, 6))
plt.pie(totales_por_categoria['Ingresos'], autopct='${:,.2f}'.format, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Distribución de Ingresos por Categoría')
plt.axis('equal')
plt.legend(totales_por_categoria['Categoría del Producto'], loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
import pandas as pd

# URLs de los archivos CSV
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer los archivos CSV en DataFrames
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular las estadísticas del costo de envío por tienda
estadisticas_costo_envio = []
for i, tienda in enumerate([tienda1, tienda2, tienda3, tienda4]):
    nombre_tienda = f'Tienda {i+1}'
    media = tienda['Costo de envío'].mean()
    desviacion_estandar = tienda['Costo de envío'].std()
    costo_envio_maximo = tienda['Costo de envío'].max()
    costo_envio_minimo = tienda['Costo de envío'].min()
    envios_cero = (tienda['Costo de envío'] == 0).sum()
    envios_no_cero = (tienda['Costo de envío'] != 0).sum()

    estadisticas_costo_envio.append({
        'Tienda': nombre_tienda,
        'Media': media,
        'Desv. Est.': desviacion_estandar,
        'Costo Máx.': costo_envio_maximo,
        'Costo Mín.': costo_envio_minimo,
        'Envios = 0': envios_cero,
        'Envios != 0': envios_no_cero
    })

# Crear un DataFrame para mostrar los resultados
resultados = pd.DataFrame(estadisticas_costo_envio)

# Formatear los números para una mejor presentación
columnas_formatear = ['Media', 'Desv. Est.', 'Costo Máx.', 'Costo Mín.']
for columna in columnas_formatear:
    resultados[columna] = resultados[columna].map('${:,.2f}'.format)

# Mostrar la tabla
print("--- Estadísticas Costo Envío por Tienda ---")
print(resultados)
```

```python
import pandas as pd
import folium

# URLs de los archivos CSV
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer los archivos CSV en DataFrames
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Crear un diccionario para mapear el nombre de la tienda a un color
tienda_colores = {
    "Tienda 1": "red",
    "Tienda 2": "blue",
    "Tienda 3": "green",
    "Tienda 4": "purple"
}

# Crear un mapa centrado en la latitud y longitud promedio de todas las ventas
all_lats = pd.concat([tienda1['lat'], tienda2['lat'], tienda3['lat'], tienda4['lat']])
all_lons = pd.concat([tienda1['lon'], tienda2['lon'], tienda3['lon'], tienda4['lon']])
center_lat = all_lats.mean()
center_lon = all_lons.mean()

m = folium.Map(location=[center_lat, center_lon], zoom_start=10)

# Función para agregar marcadores de ventas para una tienda
def add_venta_markers(tienda_df, tienda_nombre, color):
    """Agrega marcadores al mapa para cada venta de una tienda.

    Args:
        tienda_df (pd.DataFrame): DataFrame de la tienda.
        tienda_nombre (str): Nombre de la tienda.
        color (str): Color de los marcadores de la tienda.
    """
    for index, row in tienda_df.iterrows():
        lat = row['lat']
        lon = row['lon']
        folium.CircleMarker(
            location=[lat, lon],
            radius=5,  # Ajusta el tamaño de los círculos según sea necesario
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.6,
            popup=f"Venta de {tienda_nombre}"  # Puedes agregar más información al popup si lo deseas
        ).add_to(m)

# Agregar marcadores de ventas para cada tienda
add_venta_markers(tienda1, "Tienda 1", tienda_colores["Tienda 1"])
add_venta_markers(tienda2, "Tienda 2", tienda_colores["Tienda 2"])
add_venta_markers(tienda3, "Tienda 3", tienda_colores["Tienda 3"])
add_venta_markers(tienda4, "Tienda 4", tienda_colores["Tienda 4"])

# Guardar el mapa como un archivo HTML
m.save("mapa_ventas_por_tienda.html")

print("Mapa de ventas por tienda generado y guardado como mapa_ventas_por_tienda.html")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs de los archivos CSV
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer los archivos CSV en DataFrames
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Función para graficar la distribución de la ciudad de compra por tienda
def graficar_ciudad_compra(tienda_df, tienda_nombre):
    """
    Genera un gráfico de barras que muestra la distribución de la ciudad de compra para una tienda.

    Args:
        tienda_df (pd.DataFrame): DataFrame de la tienda.
        tienda_nombre (str): Nombre de la tienda.
    """

    ciudad_compra_counts = tienda_df['Lugar de Compra'].value_counts()
    ciudad_compra_df = pd.DataFrame({
        'Ciudad': ciudad_compra_counts.index,
        'Cantidad': ciudad_compra_counts.values
    })

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Ciudad', y='Cantidad', data=ciudad_compra_df, palette='viridis')
    plt.title(f'Distribución de Ciudad de Compra en {tienda_nombre}')
    plt.xlabel('Ciudad de Compra')
    plt.ylabel('Cantidad de Compras')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Graficar para cada tienda
graficar_ciudad_compra(tienda1, "Tienda 1")
graficar_ciudad_compra(tienda2, "Tienda 2")
graficar_ciudad_compra(tienda3, "Tienda 3")
graficar_ciudad_compra(tienda4, "Tienda 4")
```
